```python
import pandas as pd
```

```python
df = pd.read_csv("data\Postgres\Joined_Tables_MN.csv")
```

```python
# removing null values from the HH income zip code column that appeared after joining the two tables

df=df.dropna(axis=0, subset=['zip_code'])
```

```python
df
```

```python
df.drop(columns=['res_name', 'address', 'zip_codes', 'city', 'latitude', 'longitude', 'stdev'])
```

```python
# Displaying all food type results for cleaning

pd.set_option('display.max_rows', None)  
df['res_type'].value_counts()
```

```python
# Grouping non-restaurant results
df['res_type'] =df['res_type'].replace({"servicestations": "non_restaurant", "convenience": "non_restaurant", "catering": "non_restaurant",
            "fooddeliveryservices": "non_restaurant", "venues": "non_restaurant", "grocery": "non_restaurant", "musicvenues": "non_restaurant", 
            "services": "non_restaurant", "markets": "non_restaurant", "beverage_stores": "non_restaurant", "casinos": "non_restaurant",
                                       "butcher": "non_restaurant", "movietheaters":  "non_restaurant"})
```

```python
# Grouping American Latin food
df['res_type'] =df['res_type'].replace({"peruvian": "latin", "salvadoran": "latin", "cuban": "latin", "venezuelan": "latin",
                    "dominican": "latin", "colombian": "latin", "puertorican": "latin", "empanadas": "latin" })
```

```python
# Grouping alcohol-centric results
df['res_type'] =df['res_type'].replace({"bars": "bars/alcohol", "sportsbars": "bars/alcohol", "pubs": "bars/alcohol", 
                "cocktailbars": "bars/alcohol","gastropubs": "bars/alcohol", "divebars": "bars/alcohol","beer_and_wine": "bars/alcohol", 
                "lounges": "bars/alcohol", "beerbar": "bars/alcohol", "wine_bars": "bars/alcohol", "breweries": "bars/alcohol", 
                "gaybars": "bars/alcohol", "irish_pubs": "bars/alcohol", "brewpubs": "bars/alcohol","beergardens": "bars/alcohol" })
```

```python
# Grouping middle eastern results
df['res_type'] =df['res_type'].replace({"halal": 'mideastern', "falafel": "mideastern", "afghani": "mideastern", "moroccan": "mideastern",
                    "lebanese": "mideastern", "persian": "mideastern", "turkish": 'mideastern', "kebab": "mideastern"})
```

```python
# Grouping similar foods
df['res_type'] =df['res_type'].replace({"newamerican": "american", "tradamerican": "american", "desserts": "bakeries", "donuts": "bakeries", 
                "bagels": "bakeries", "sandwiches": 'delis/diners', "delis": 'delis/diners', 'diners':'delis/diners', "hotdog": 'hotdogs',
                "chicken_wings": "wings/chicken", "chickenshop": "wings/chicken", "waffles": "breakfast_brunch",
                "mexican":"mex/tex-mex", "tex-mex":"mex/tex-mex", "tacos": "mex/tex-mex" })
```

```python
# Grouping diet needs
df['res_type'] =df['res_type'].replace({'gluten-free': "special diet", "vegan": "special diet", 'vegetarian': "special diet"})
```
