```python
import pandas as pd
import glob
import os
```

```python
# setting the path for joining multiple files
files = os.path.join("C:\\Users\\menon\\msu\\Data_Analytics_Class\\Final Project\\FastFoodFinal\\data\\Yelp_Results\\", "State_*.csv")

# list of merged files returned
files = glob.glob(files)

print("Single CSV of All Result Files");

# joining files with concat and read_csv
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
print(df)
```

```python
df
```

```python
df= df.astype({'Zip Code': 'string', "Address": "string"})
```

```python
df.dtypes
```

```python
df['Zip Code'] = df['Zip Code'].str.replace(r'\D', '')
df['Address'] = df['Address'].str.replace(r'\W', '')
```

```python
df
```

```python
df.to_csv('data/All_Yelp_Results.csv', columns = ["Name","Type", "Address", "Zip Code"], header = True, index= False)
```
